Create a new road with 4 lanes and a few vehicles

```python
from highway.road.road import Road

road = Road.create_random_road(lanes_count=4, lane_width=4.0, vehicles_count=3)
```

Step the road state of 1 second

```python
road.step(dt=1.0)
print(road)
```

Create a *Simulation* from the created road

```python
from highway.simulation import Simulation
from highway.agent.ttc_vi import TTCVIAgent
from highway.vehicle.control import MDPVehicle

sim = Simulation(road, ego_vehicle_type=MDPVehicle, agent_type=TTCVIAgent)

sim.process()
sim.quit()
```
