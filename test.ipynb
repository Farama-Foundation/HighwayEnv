```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import matplotlib.pyplot as plt

import gymnasium
import highway_env
highway_env.register_highway_envs()

env = gymnasium.make("parking-v0", render_mode='rgb_array')
env.configure({
    "vehicles_count": 10, 
    "duration": 300,
    "screen_width": 900,
    "screen_height": 1200,
    "collision_reward": -8,
    "custom_reward": False,
    "n_rows": 4,
    "y_offset": 10, # y distance between parallel rows
    "obstacles": True,
    "random_start": True
})
env.reset()

fig,ax = plt.subplots(1, figsize=(10,10))
ax.imshow(env.render())
env.close()
```
